```python
import matplotlib.pyplot as plt
import numpy as np

tiempo = np.arange(0, 10, 0.1)
N = len(tiempo)
```

- **loc**: Mean of the distribution
- **scale**: Standard deviation of the distribution
- **size**: Sample size

```python
x = np.random.normal(loc=2, scale=0.5, size=N)
plt.plot(tiempo, x)
plt.axhline(np.mean(x), linestyle='--', color='red')
plt.axhline(y=np.mean(x) + 2 * np.std(x), linestyle=':', color='black')
plt.axhline(y=np.mean(x) - 2 * np.std(x), linestyle=':', color='black')
```

```python
# Histograma
hx, bins, _ = plt.hist(x, bins=30, alpha=0.5, edgecolor='black', density=True)

# Media y desviación estándar
mean_x = np.mean(x)
std_x = np.std(x)

# Líneas en la media y en ±2 desviaciones estándar
plt.axvline(mean_x, linestyle='dashed', color='black', label='Media')
plt.axvline(mean_x + 2 * std_x, linestyle='dotted', color='blue', label='+2σ')
plt.axvline(mean_x - 2 * std_x, linestyle='dotted', color='blue', label='-2σ')

# Curva de distribución normal
ejex = np.linspace(min(x), max(x), 100)
fx_normal = (1 / (std_x * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((ejex - mean_x) / std_x) ** 2)
plt.plot(ejex, fx_normal, color='red', label='Distribución Normal')

# Mostrar gráfico
plt.legend()
plt.show()
```
