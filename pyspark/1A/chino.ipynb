```python
import json

with open('CHINO_2023.txt') as user_file:
  file_contents = user_file.read()
  
print(file_contents)
```

```python
parsed_json = json.loads(file_contents)
parsed_json
```

```python
print('idClase','nombre','moneda','fecha', sep='|')
print('-'*30)
registros = []
for e in parsed_json['records']:
    t = e['idClase'] + '|' + e['nombre'] + '|' + e['moneda'] + '|' + e['fecha'] + '\n'
    registros.append(t)
```

```python
f = open("archivo_nuevo.txt", "w")
f.writelines(registros)
f.close()
```
