## JSON Plano to CSV

Nos llego un archivo json en texto plano (es decir todo en una sola linea) y deseamos convertirlo en un csv

- Python 3.8.8
- Pandas 1.3.4

```python
import json
import pandas as pd

with open('filejsonplano.csv') as f:
  file_contents = f.read()
```

```python
parsed_json = json.loads(file_contents)
#  ---------------->   list json
df = pd.json_normalize(parsed_json['records'])
```

```python
# dataframe view
df
```

```python
# save to csv
df.to_csv('test.csv', index=False, header=False, sep='|', encoding='utf-8')
```
