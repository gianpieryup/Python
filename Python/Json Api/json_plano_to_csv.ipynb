## Example 1 
Nos llego un archivo json en texto plano (es decir todo en una sola linea) y deseamos convertirlo en un csv

- Python 3.8.8
- Pandas 1.3.4

```python
import json
import pandas as pd

with open('example1.csv') as f:
  file_contents = f.read()
```

```python
parsed_json = json.loads(file_contents)
#  ---------------->   list json esta dentro de la clave 'parsed_json'
df = pd.json_normalize(parsed_json['records'])
```

```python
# dataframe view
df
```

```python
df.info()
```

```python
# save to csv
df.to_csv('result.csv', index=False, header=False, sep='|', encoding='utf-8')
```

## Example 4

Archivo json, donde cada registro esta separado por un caracter separador (`;`), salto de linea seria (`\n`)

```python
import json
import pandas as pd
import re

with open('example4.csv') as f:
  file_contents = f.read()
```

```python
df_json = file_contents.split(';')
df_json
```

```python
df = pd.DataFrame(df_json, columns=['json'])
df
```

```python
# Esto lo desarma a primer nivel, me parece ideal para una primera version tabular
df_final = df.apply(lambda x: pd.Series(json.loads(x['json'])), axis=1, result_type='expand')
df_final
```

```python
df_final.info()
```

```python
# save to csv
df_final.to_csv('result.csv', index=False, header=False, sep='|', encoding='utf-8')
```
